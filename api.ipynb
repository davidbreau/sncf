```python
import requests
```

```python
params = {
    'q' : 'Paris',
    'dataset': ['objets-trouves-restitution'],
    'sort': ['date'],
    'facet': ['date', 'gc_obo_gare_origine_r_name',
    'gc_obo_nature_c', 'gc_obo_type_c'],
    'refine.date': ['2022'],
    # 'refine.gc_obo_gare_origine_r_name': ['Paris Gare de Lyon']
    }


url = "https://ressources.data.sncf.com/api/records/1.0/search/?"

response = requests.get(url, params=params)

if response.ok:
    data = response.json()
    print(data)
else:
    print("La requête a échoué avec le code d'état ", response.status_code)
```

```python
from urllib.parse import urlparse, parse_qs

url = "https://ressources.data.sncf.com/api/records/1.0/search/?dataset=objets-trouves-restitution&q=&sort=date&facet=date&facet=gc_obo_gare_origine_r_name&facet=gc_obo_nature_c&facet=gc_obo_type_c&refine.date=2022&refine.gc_obo_gare_origine_r_name=Paris+Gare+de+Lyon"

parsed_url = urlparse(url)
query_params = parse_qs(parsed_url.query)

print(query_params)
```
